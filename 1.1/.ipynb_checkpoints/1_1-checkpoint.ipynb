## Экспериментальное определение постоянной Планка

```python
%matplotlib inline 
#Импротируем необходимые функции
import numpy as np # работа с массивами и линейной алгеброй
import matplotlib.pyplot as plt # для отрисовки графиков
import pandas as pd # для чтения и работы с данными
from scipy.optimize import curve_fit # фитирующая процедура
```

```python
def line(x, k, c):
    return k * x + c
```

## Градуировка монохроматора.

```python
data = pd.read_csv('data/grad.csv', sep = ',', encoding = 'latin1')
y = data.length
x = data.angle

plt.plot(x, y, label = 'graduation') #Обычный график из полученных данных.
A = np.vstack([x, np.ones(len(x))]).T
m, b = np.linalg.lstsq(A, y)[0]
print(m)

plt.xlabel('angle, degree')
plt.ylabel('length, A')
plt.legend(loc = 0)
```

### Найдем линейную зависимость.

```python
x1 = x[0:15]
y1 = y[0:15]
res = curve_fit(f = line, xdata = x1, ydata = y1)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'Fitting')
plt.plot(x1, y1, label = 'graduation') #Обычный график из полученных данных.

plt.xlabel('angle, degree')
plt.ylabel('length, A')
plt.legend(loc = 0)

print(popt[0])
print(popt[1])

fi1 = popt[0] * 2372 + popt[1]
fi2 = popt[0] * 2125 + popt[1]
fi3 = popt[0] * 2050 + popt[1]
print(fi1) 
print(fi2)
print(fi3)
#2.2623223706
```

## Исследование зависимости показаний тока от величины тормзящего потенциала.

### Для длины волны, равной 6358,19 ангстрем.

```python
#2372 градуса на барабане fi1 = 6358.19
data = pd.read_csv('data/2372.csv', sep = ',', encoding = 'latin1')
x = data.Uf
y = data.Ui
plt.plot(x, y ** 0.5, label = 'photocurrent') #Обычный график из полученных данных.

plt.xlabel('Voltage, mV')
plt.ylabel('Current, mV')
plt.legend(loc = 0)
```

Найдем величину запирающего потенциала.

```python
#Подбираем прямую
x1 = x[7:23]
y1 = y[7:23]
res = curve_fit(f = line, xdata = x1, ydata = y1 **0.5)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'Fitting')
plt.plot(x1, y1 ** 0.5, label = 'photocurrent') #Обычный график из полученных данных.

plt.xlabel('Voltage, mV')
plt.ylabel('Current, mV')
plt.legend(loc = 0)

print(popt[0])
print(popt[1]) 
print('Величина запирающего потенциала для длины волны 6358.19 ангстрем:',-popt[1]/popt[0])
```

## Построим график зависимости величины запирающего напряжения от частоты падающего излучения.

```python
data = pd.read_csv('data/result2.csv', sep = ',', encoding = 'latin1')
x = data.frequency
y = data.voltage
plt.plot(3/x * 10**6 , y, label = 'result') #Обычный график из полученных данных.

plt.xlabel('frequency, THz')
plt.ylabel('voltage, mV')
plt.legend(loc = 0)
```

## По графику определим постоянную Планка по формуле: $\frac{dV_0}{d\omega} = \frac{\hbar}{e}$

```python
#Подбираем прямую
x1 = 3 * 10**6 / x * 2 *3.14
y1 = y
res = curve_fit(f = line, xdata = x1, ydata = y1)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'Fitting')
plt.plot(x1, y1, label = 'result') #Обычный график из полученных данных.

plt.xlabel('frequency, THz')
plt.ylabel('voltage, mV')
plt.legend(loc = 0)

A = np.vstack([x1, np.ones(len(x1))]).T
m, b = np.linalg.lstsq(A, y1)[0]
print(m)
print('h =', m * 1.6 * 10**(-34))

print('h =', popt[0] * 1.6 * 10**(-34))
```

## Выводы: 
Полученная постоянная Планка -- 0.66е-34.

Табличное значение постонной Планка -- 1.05е-34.
