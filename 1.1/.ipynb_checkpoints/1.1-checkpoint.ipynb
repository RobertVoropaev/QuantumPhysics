## Экспериментальное определение постоянной Планка

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
from scipy.optimize import curve_fit
```

```python
def line(x, a, b):
    return a * x + b
```

## Градуировка монохроматора.

### Найдем линейную зависимость.

```python
data = pd.read_csv('data/grad.csv', sep = ',')
y = data.length
x = data.angle

x1 = x.copy()
y1 = y.copy()
res = curve_fit(f = line, xdata = x1, ydata = y1)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'fitting')
plt.plot(x1, y1, label = 'graduation')

plt.xlabel('angle, degree')
plt.ylabel('length, A')
plt.legend(loc = 0)
plt.show()

print(*popt)

fi1 = line(2372, *popt)
fi2 = line(2125, *popt)
fi3 = line(2050, *popt)
print(fi1) 
print(fi2)
print(fi3)
```

## Исследование зависимости показаний тока от величины тормзящего потенциала.

### Для длины волны, равной 6358,19 ангстрем.

```python
#2372 градуса на барабане fi1 = 6358.19
data = pd.read_csv('data/2372.csv', sep = ',', encoding = 'latin1')
x = data.Uf
y = data.Ui
plt.plot(x, y ** 0.5, label = 'photocurrent') 

plt.xlabel('Voltage, mV')
plt.ylabel('Current, mV')
plt.legend(loc = 0)

plt.show()
```

Найдем величину запирающего потенциала.

```python
#Подбираем прямую
x1 = x[7:23]
y1 = y[7:23]
res = curve_fit(f = line, xdata = x1, ydata = y1 **0.5)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'Fitting')
plt.plot(x1, y1 ** 0.5, label = 'photocurrent') 

plt.xlabel('Voltage, mV')
plt.ylabel('Current, mV')
plt.legend(loc = 0)

print(*popt)
print('Величина запирающего потенциала для длины волны 6358.19 ангстрем:',-popt[1]/popt[0])

plt.show()
```

## Построим график зависимости величины запирающего напряжения от частоты падающего излучения.

```python
data = pd.read_csv('data/result2.csv', sep = ',', encoding = 'latin1')
x = data.frequency
y = data.voltage
plt.plot(3/x * 10**6 , y, label = 'result') 

plt.xlabel('frequency, THz')
plt.ylabel('voltage, mV')
plt.legend(loc = 0)

plt.show()
```

## По графику определим постоянную Планка по формуле: $\frac{dV_0}{d\omega} = \frac{\hbar}{e}$

```python
#Подбираем прямую
x1 = 3 * 10**6 / x * 2 *3.14
y1 = y
res = curve_fit(f = line, xdata = x1, ydata = y1)
popt, pcov = res
plt.plot(x1, line(x1, *popt), label = 'Fitting')
plt.plot(x1, y1, label = 'result') 

plt.xlabel('frequency, THz')
plt.ylabel('voltage, mV')
plt.legend(loc = 0)

res = curve_fit(f = line, xdata = x1, ydata = y1)
popt, pcov = res
m, b = popt
print(m)
print('h =', m * 1.6 * 10**(-34))

plt.show()
```

## Выводы: 
Полученная постоянная Планка -- 0.66е-34.

Табличное значение постонной Планка -- 1.05е-34.
